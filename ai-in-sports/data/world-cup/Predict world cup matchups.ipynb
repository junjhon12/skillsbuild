```python

import os, types
import pandas as pd
from botocore.client import Config
import ibm_boto3

def __iter__(self): return 0

# @hidden_cell
# The following code accesses a file in your IBM Cloud Object Storage. It includes your credentials.
# You might want to remove those credentials before you share the notebook.

cos_client = ibm_boto3.client(service_name='s3',
    ibm_api_key_id='LcyS2t8qRdTQ1ket25gYggmKTGFhQ85mspd0bjMLfF3k',
    ibm_auth_endpoint="https://iam.cloud.ibm.com/identity/token",
    config=Config(signature_version='oauth'),
    endpoint_url='https://s3.direct.us-south.cloud-object-storage.appdomain.cloud')

bucket = 'sandboxproject-donotdelete-pr-snjjamdlq8h7ju'
object_key = 'player_latest_market_value.csv'

body = cos_client.get_object(Bucket=bucket,Key=object_key)['Body']
# add missing __iter__ method, so pandas accepts body as file-like object
if not hasattr(body, "__iter__"): body.__iter__ = types.MethodType( __iter__, body )

player_values = pd.read_csv(body)
player_values.head(10)
```

```python
players_merged_values = pd.merge(
    players, 
    player_values, 
    on='player_id',
    how='left'
)

players_merged_values.head(10)
```

```python

# 1. Convert the 'date_of_birth' column to datetime objects
# errors='coerce' will turn invalid dates into NaT (Not a Time) instead of crashing
players_merged_values['date_of_birth'] = pd.to_datetime(players_merged_values['date_of_birth'], errors='coerce')

# 2. Get today's date (or set a specific date like the World Cup start)
now = pd.Timestamp.now()

# 3. Calculate age
# We subtract the dates to get 'Timedeltas', then convert to days and divide by 365.25
players_merged_values['age'] = (now - players_merged_values['date_of_birth']).dt.days / 365.25

# 4. Optional: Round down to the nearest whole number
players_merged_values['age'] = players_merged_values['age'].fillna(0).astype(int)

print(players_merged_values[['player_name', 'date_of_birth', 'age']].head())
```

```python
# Merge them (Left Join keeps all players)
# This results in one big table with player names, values, and their club names
df_complete = pd.merge(
    players_merged_values, 
    teams[['club_id', 'club_name', 'competition_id', 'club_division']], 
    left_on='current_club_id', 
    right_on='club_id', 
    how='left'
)

# Create the National Team Power Ranking
# We group by the player's country and sum their market values
world_cup_ranking = df_complete.groupby('citizenship')['value'].sum().reset_index()

# Sort to see the top contenders
world_cup_ranking = world_cup_ranking.sort_values(by='value', ascending=False)

# 3. Format the numbers to be more readable (in Millions)
world_cup_ranking['Value in Millions'] = world_cup_ranking['value'] / 1_000_000

# 3. Sort by the new 'Team Name' column
teams_current_sorted = df_complete.sort_values(by='club_name')

print(world_cup_ranking.head(10))
```

```python
def predict_match(team_a_name, team_b_name, ranking_df):
    # 1. Get the values for both teams
    try:
        val_a = ranking_df.loc[ranking_df['citizenship'] == team_a_name, 'value'].values[0]
        val_b = ranking_df.loc[ranking_df['citizenship'] == team_b_name, 'value'].values[0]
    except IndexError:
        return "One of the teams was not found in the dataset."

    # 2. Calculate the Ratio
    ratio = val_a / val_b if val_b > 0 else 100
    
    # 3. Assign Logic for Wins
    if ratio > 2.0:
        prediction = f"{team_a_name} is a Heavy Favorite"
        prob = "75-90%"
    elif 1.2 < ratio <= 2.0:
        prediction = f"{team_a_name} is the Favorite"
        prob = "60-70%"
    elif 0.8 <= ratio <= 1.2:
        prediction = "It's a Toss-up / Likely Draw"
        prob = "50/50"
    elif 0.5 <= ratio < 0.8:
        prediction = f"{team_b_name} is the Favorite"
        prob = "60-70%"
    else:
        prediction = f"{team_b_name} is a Heavy Favorite"
        prob = "75-90%"

    return f"Match: {team_a_name} vs {team_b_name} | Result: {prediction} (Confidence: {prob})"

# Example Usage:
result = predict_match('France', 'Morocco', world_cup_ranking)
print(result)
```

```python
# List of fixtures
fixtures = [('France', 'Poland'), ('England', 'Senegal'), ('Argentina', 'Australia')]

for team_a, team_b in fixtures:
    print(predict_match(team_a, team_b, world_cup_ranking))
```

```python
# 1. Prepare a unique Country-Level lookup from your power_ranking data
# We take the unique 'citizenship' and its 'squad_total_value'
country_lookup = power_ranking[['citizenship', 'squad_total_value']].drop_duplicates()

# 2. First Merge: Attach Home Team Value
df_with_home = pd.merge(
    results, 
    country_lookup, 
    left_on='home_team', 
    right_on='citizenship', 
    how='left'
)

# Rename to distinguish from away team and drop the redundant join key
df_with_home = df_with_home.rename(columns={'squad_total_value': 'home_squad_value'})
df_with_home = df_with_home.drop(columns=['citizenship'])

# 3. Second Merge: Attach Away Team Value
df_final = pd.merge(
    df_with_home, 
    country_lookup, 
    left_on='away_team', 
    right_on='citizenship', 
    how='left'
)

# Rename and cleanup
df_final = df_final.rename(columns={'squad_total_value': 'away_squad_value'})
df_final = df_final.drop(columns=['citizenship'])

# 4. Create the Target Column for AutoAI
def get_result(row):
    if row['home_score'] > row['away_score']: return 'Home Win'
    if row['home_score'] < row['away_score']: return 'Away Win'
    return 'Draw'

df_final['target_result'] = df_final.apply(get_result, axis=1)

# 5. Clean up missing values
# If a country in results.csv wasn't in your Transfermarkt data, set value to 0
df_final[['home_squad_value', 'away_squad_value']] = df_final[['home_squad_value', 'away_squad_value']].fillna(0)

print(df_final[['home_team', 'away_team', 'home_squad_value', 'away_squad_value', 'target_result']].head())
```
